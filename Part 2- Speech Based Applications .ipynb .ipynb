### **Deploy Application Demo with Gradio**

```python
import os
import gradio as gr
demo = gr.Blocks()
```

```python
def transcribe_speech(filepath):  # to upload audio files
    if filepath is None:          # make sure there is sound
        gr.Warning("No audio found, please retry.")
        return ""

    output = bot(filepath)
    return output["text"]
```

**User interface**

```python
mic_transcribe = gr.Interface(
    fn = transcribe_speech,
    inputs=gr.Audio(sources="microphone",
                    type="filepath"),
    outputs=gr.Textbox(label="Transcription",
                       lines=3),
    allow_flagging="never")
```

**User interface to Recoding in real time**

```python
file_transcribe = gr.Interface(
    fn = transcribe_speech,
    inputs = gr.Audio(sources="upload",
                    type="filepath"),
    outputs = gr.Textbox(label="Transcription",
                       lines=3),
    allow_flagging = "never",
)
```

```python
import gradio as gr

# Gradio app
demo = gr.TabbedInterface(
    [mic_transcribe, file_transcribe],
    ["Transcribe Microphone", "Transcribe Audio File"], # User interface to upload files
)

demo.launch()
```
